```python
# support functions

def results_from_path(path: Optional[str], experiment_name: Optional[str] = None, date: Optional[str] = None):
    if path is None and experiment_name is None and date is None:
        raise ValueError
    with open(path, 'rb') as file:
        res = pickle.load(file)
    return res

def merge_results(*results_list):
    n = len(results_list)
    merged_res = [None]*5

    for i in range(5):
        running_list = []

        for r in range(n):
            running_list = running_list + results_list[r][i]
        merged_res[i] = running_list
    return merged_res
```

```python
def extract_reward_mean(res):
    rew_mean = []
    for i in range(len(res)):
        rew_mean.append(res[i]["rewards"].mean())
    return rew_mean

def extract_turn_mean(res):
    turn_mean = []
    for i in range(len(res)):
        turn_mean.append(res[i]["game_turns"].mean())
    return turn_mean
```

```python
def plot_rew(results, max=900, labels=None, max_x = 400):
    """
    Plot mean rewards for three result objects a, b, c.
    Each of a, b, c must be compatible with extract_reward_mean(res).
    """
    a, b, c = [], [], []
    
    for res in results:
        a.append(res[2])
        b.append(res[3])
        c.append(res[4])
    
    fig, axs = plt.subplots(1, 3, figsize=(16, 6), sharey=True)

    def add_plot(ax, res, title, label=None):
        series = np.asarray(extract_reward_mean(res)[: max], dtype=float)
        x = np.arange(len(series))[:400]
        if label is None:
            ax.plot(x, series, label="mean reward")
        else:
            ax.plot(x, series, label=label)
        ax.set_title(title)
        ax.set_xlabel("iteration")
        ax.grid(True, alpha=0.3)
        ax.legend()

    for i in range(len(a)):
        add_plot(axs[0], a[i], "Agent 1 VS Agent 2", labels[i])
        add_plot(axs[1], b[i], "Agent 1 VS Random", labels[i])
        add_plot(axs[2], c[i], "Agent 2 VS Random", labels[i])

    plt.ylim(-1, 1) 

    axs[0].set_ylabel("mean reward")
    plt.tight_layout()
    plt.show()
```

```python
def plot_turn(results, max=900, labels=None):
    """
    Plot mean rewards for three result objects a, b, c.
    Each of a, b, c must be compatible with extract_reward_mean(res).
    """
    a, b, c = [], [], []
    
    for res in results:
        a.append(res[2])
        b.append(res[3])
        c.append(res[4])
    
    fig, axs = plt.subplots(1, 3, figsize=(16, 6), sharey=True)

    def add_plot(ax, res, title, label=None):
        series = np.asarray(extract_turn_mean(res)[: max], dtype=float)
        x = np.arange(len(series))
        if label is None:
            ax.plot(x, series, label="mean turn")
        else:
            ax.plot(x, series, label=label)
        ax.set_title(title)
        ax.set_xlabel("iteration")
        ax.grid(True, alpha=0.3)
        ax.legend()

    for i in range(len(a)):
        add_plot(axs[0], a[i], "Agent 1 VS Agent 2", labels[i])
        add_plot(axs[1], b[i], "Agent 1 VS Random", labels[i])
        add_plot(axs[2], c[i], "Agent 2 VS Random", labels[i])

    plt.ylim(40, 70) 

    axs[0].set_ylabel("mean turn")
    plt.show()
```

```python
def plot_decision(output: Tensor | np.ndarray, log: bool = True):
    if isinstance(output, Tensor):
        im = output.cpu().detach().numpy()
    elif isinstance(output, np.ndarray):
        im = output
    else:
        raise TypeError("Input must be Tensor or ndarray")

    im = im.squeeze()

    if im.shape == (81,):
        im = im.reshape(9, 9)
    elif im.shape != (9, 9):
        raise ValueError("Input must have shape (9,9) or (81,)")

    fig, ax = plt.subplots()
    img = ax.imshow(im, cmap='gray')

    # Add probabilities in each cell
    for i in range(9):
        for j in range(9):
            value = im[i, j]
            ax.text(
                j, i,
                f"{value:.2f}",
                ha='center',
                va='center',
                color='white' if value < im.max()/2 else 'black',
                fontsize=8
            )

    # Colorbar
    cbar = fig.colorbar(img, ax=ax)
    cbar.set_label('Probability', rotation=270, labelpad=15)

    # Thick 3×3 grid
    for x in range(0, 10, 3):
        ax.axvline(x - 0.5, color='red', linewidth=2)
        ax.axhline(x - 0.5, color='red', linewidth=2)

    # Fine grid
    ax.set_xticks(np.arange(-0.5, 9, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 9, 1), minor=True)
    ax.grid(which='minor', color='gray', linewidth=0.5)

    ax.tick_params(which='both', bottom=False, left=False,
                   labelbottom=False, labelleft=False)

    plt.show()
```
